```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
data=pd.read_csv('iris.csv')
df=pd.DataFrame(data)
x=df[0,2,4]
x=np.matrix(x)
y=df[5]
y=y.map({'Iris-setosa':0,'Iris-versicolor':1})
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=50,test_size=0.3)
print(x_train)
```

```python
epochs=1000
alpha=0.5
##Training the data
weight=np.ones((4,1))
bias=0
maxaccuracy=0
accuracy={}

def calculate(x,weight,b):
    if(np.dot(x,weight)>=b):
        return 1
    else:
        return 0

for j in range(1,epochs+1):
    for i in range(0,len(x_train)):
        ypred=calculate(x_train[i],weight,bias)
        if(ypred==1 and np.array(y_train)[i]==0):
            weight+=alpha*np.transpose(x_train[i])
            bias-=alpha*1
            print("done")
        elif(ypred==0 and np.array(y_train)[i]==1):
            weight-=alpha*np.transpose(x_train[i])
            bias+=alpha*1
            print("not done")
    
    tempy=[]
    for i in range(0,len(x_train)):
            y=calculate(x_train[i],weight,bias)
            tempy.append(y)
    tempy=np.array(tempy)
    accuracy[j]=accuracy_score(tempy,y_train)
    if(accuracy[j]>maxaccuracy):
        maxaccuracy=accuracy[j]
        markweight=weight
        markbias=bias

plt.plot(accuracy.values())
plt.xlabel('iterations')
plt.ylabel('Accuracy')
plt.ylim([0,1])
plt.show()
print("final weight vector is",weight)
    
    
```
