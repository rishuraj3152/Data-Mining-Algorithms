```python
dt = pd.read_csv('Absenteeism_at_work.csv')
new = dt["Work load Average"].str.split(",", n = 1, expand = True)  
dt['a'] = pd.to_numeric(new[0], errors='coerce')
dt['b'] = pd.to_numeric(new[1], errors='coerce')
dt['Work Average'] = pd.DataFrame((dt[['a']].values / dt[['b']].values)*10)
dt['Work Average'] = dt['Work Average'].astype(int)
dt['Work Average']
```

```python
f1 = dt['Reason for absence'].values
f2 = dt['Month of absence'].values
f3 = dt['Day of the week'].values
f4 = dt['Seasons'].values
f5 = dt['Transportation expense'].values
f6 = dt['Distance from Residence to Work'].values
f7 = dt['Service time'].values
f8 = dt['Age'].values
f9 = dt['Hit target'].values
f10 = dt['Disciplinary failure'].values
f11 = dt['Education'].values
f12 = dt['Son'].values
f13 = dt['Social drinker'].values
f14 = dt['Social smoker'].values
f15 = dt['Pet'].values
f16 = dt['Weight'].values
f17 = dt['Height'].values
f18 = dt['Body mass index'].values
f19 = dt['Absenteeism time in hours'].values
f20 = dt['Work Average'].values
```

```python
data = np.array(list(zip(f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12, f13, f14, f15, f16, f17, f18, f19,f20,)))
D = pairwise_distances(data, metric='euclidean')
```

```python
k = 5
```

```python
tmax = 100
```

```python
m, n = D.shape
```

```python
if k > n :
    raise Exception('too many medoids')
```

```python
valid_medoid_inds = set(range(n))
invalid_medoid_inds = set([])
rs, cs = np.where(D==0)
print(valid_medoid_inds)
print(invalid_medoid_inds)
print(rs)
print(cs)
```

```python
index_shuf = list(range(len(rs)))
np.random.shuffle(index_shuf)
rs = rs[index_shuf]
cs = cs[index_shuf]
```

```python
for r, c in zip(rs, cs):
    if r < c and r not in invalid_medoid_inds:
        invalid_medoid_inds.add(c)
valid_medoid_inds = list(a - b for a, b in zip(valid_medoid_inds, invalid_medoid_inds))
```

```python
if k > len(valid_medoid_inds):
    raise Exception('Too many medoids (After removing {} duplicate points)'.format(len(invalid_medoid_inds)))
```

```python
M = np.array(valid_medoid_inds)
np.random.shuffle(M)
M = np.sort(M[:k])
```

```python
Mnew = np. copy(M)
```

```python
C = {}
for t in range(tmax):
    J = np.argmin(D[:, M], axis = 1)
    for i in range(k):
        C[i] = np.where(J == i)[0]
    for i in range(k):
        J = np.mean(D[np.ix_(C[i], C[i])],axis=1)
        j = np.argmin(J)
        Mnew[i] = C[i][j]
    np.sort(Mnew)
    if np.array_equal(M, Mnew):
        break
    M = np.copy(Mnew)
else:
    J = np.argmin(D[:, M], axis = 1)
    for i in range(k):
        C[i] = np.where(J==i)[0]
```

```python
print('MEDOIDS:')
for i in M:
    print(data[i])
```

```python
print('Clustering Result:')
for label in C:
    for i in C[label]:
        print('Label {0}:{1}'. format(label, data[i]))
```
