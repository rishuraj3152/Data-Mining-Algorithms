```python
import numpy as np
import pandas as pd
import random
data=pd.read_excel('Absenteeism_at_work.xls')
##generating two random number for making clusters
first=random.randrange(0,739,1)
second=random.randrange(0,739,1)
print("First=",first,"Second=",second)

##considering k=2
data.head()
firstmedoid=np.array(data[first:first+1])
secondmedoid=np.array(data[second:second+1])
#print(firstmedoid,secondmedoid)
m,n=data.shape
cluster1=[]
cluster2=[]
totalcost=0
for i in range(0,m-1):
    if(i!=first and i!=second):
        dist1=np.sum(np.abs(firstmedoid-np.array(data[i:i+1])))
        dist2=np.sum(np.abs(secondmedoid-np.array(data[i:i+1])))
        if(dist1<dist2):
            cluster1.append(data[i:i+1])
            totalcost+=dist1
        else:
            cluster2.append(data[i:i+1])
            totalcost+=dist2
print("Total Cost =",totalcost)

##consider replacing first medoid with other medoids
for k in range(m):
    if(k!=second):
        temp=k
    else:
        temp=k+1
#temp=random.randrange(0,739,1)
    optionalmedoid=np.array(data[temp:temp+1])
    tempcluster1=[]
    tempcluster2=[]
    temptotalcost=0
    for i in range(0,m-1):
        if(i!=temp and i!=second):
            dist1=np.sum(np.abs(optionalmedoid-np.array(data[i:i+1])))
            dist2=np.sum(np.abs(secondmedoid-np.array(data[i:i+1])))
            if(dist1<dist2):
                tempcluster1.append(data[i:i+1])
                temptotalcost+=dist1
            else:
                tempcluster2.append(data[i:i+1])
                temptotalcost+=dist2

    print("Temp TotalCost =",temptotalcost)
    if(temptotalcost<totalcost):
        print("the new clustering is better")
    else:
        print("previous clustering is better")
    
##consider replacing second medoid with other medoids
for k in range(m):
    if(k!=second):
        temp=k
    else:
        temp=k+1
    optionalmedoid=np.array(data[temp:temp+1])
    tempcluster1=[]
    tempcluster2=[]
    temptotalcost=0
    for i in range(0,m-1):
        if(i!=temp and i!=second):
            dist1=np.sum(np.abs(optionalmedoid-np.array(data[i:i+1])))
            dist2=np.sum(np.abs(secondmedoid-np.array(data[i:i+1])))
            if(dist1<dist2):
                tempcluster1.append(data[i:i+1])
                temptotalcost+=dist1
            else:
                tempcluster2.append(data[i:i+1])
                temptotalcost+=dist2

    print("Temp TotalCost =",temptotalcost)
    if(temptotalcost<totalcost):
        print("the new clustering is better")
    else:
        print("previous clustering is better")
```
