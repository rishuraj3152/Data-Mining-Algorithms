```python
import numpy as np
import pandas as pd
import random as rd

data=pd.read_excel('data1.xlsx')
m,n=data.shape
print(data.shape)
visited=np.zeros(m,dtype='int32')
data['visited']=0
data.head()
```

```python
def update_reachability_dis(y,x,epsilon):
    return max(euclideandis(x,y),epsilon)
    

def euclideandis(x,y):
    return np.sqrt((x[0]-y[0])*(x[0]-y[0])+(x[1]-y[1])*(x[1]-y[1]))

distancematrix=np.empty((m,m))
print(distancematrix.shape)
label,freq=np.unique(data['visited'],return_counts=True)
print(label,freq[0])

epsilon=0.02
minpts=150
random=rd.randrange(0,m-1,1)
data.iloc[random][3]=1
orderseeds=[random]
for i in range(len(orderseeds)):
    ##calculating core distance
    dis=[]
    for i in range(0,m):
        if(i!=random):
            point=data.iloc[i]
            dis.append(euclideandis(point,data.iloc[random]))
    dis.sort()
    ##considering minpts value to be 150
    smallestepsilon=dis[149]

    print("core distance of point p is",smallestepsilon)
    for i in range(0,m):
        if(i!=random):
            point=data.iloc[i]
            if(euclideandis(point,data.iloc[random])<smallestepsilon and data.iloc[i][3]==0):
                    data.iloc[i][3]=1
                    orderseeds.append(i)
                    distancematrix[random][i]=update_reachability_dis(point,data.iloc[random],smallestepsilon)
        print("reachability distance of points lying in the neighbourhood of p",distancematrix)
```
