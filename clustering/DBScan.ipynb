```python
import numpy as np
import pandas as pd
import random as rd

data=pd.read_excel('data1.xlsx')
m,n=data.shape
print(data.shape)
visited=np.zeros(m,dtype='int32')
data['visited']=0
data.head()
```

```python
def euclideandis(x,y):
    return np.sqrt((x[0]-y[0])*(x[0]-y[0])+(x[1]-y[1])*(x[1]-y[1]))


epsilon=1
minpts=10
label,freq=np.unique(data['visited'],return_counts=True)
print(label,freq[0])
while(freq[0]>0):
    r=rd.randrange(0,m-1,1)
    center=data.iloc[r]
    data.iloc[r][3]=1
    #area=3.14*epsilon*epsilon
    count=0
    neighbourhood=[]
    for i in range(m):
        if(i!=r):
            point=data.iloc[i]
            dis=euclideandis(point,center)
            if(dis<epsilon):
                count+=1
                neighbourhood.append(i)
    if(count>=minpts):
        cluster=[]
        cluster.append(r)
        for j in neighbourhood:
            if(data.iloc[j][3]==0 and j!=r):
                data.iloc[j][3]=1
                count1=0
                tempneighbour=[]
                for k in range(m):
                    point=data.iloc[k]
                    dis=euclideandis(point,data.iloc[j])
                    if(dis<epsilon):
                        count1+=1
                        tempneighbour.append(k)
                if(count1>=minpts):
                    for iter in tempneighbour:
                        if(iter not in neighbourhood):
                            neighbourhood.append(iter)
            if(j not in cluster):
                cluster.append(j)
        print("cluster formed have following points",cluster)
    else:
        print(r,"is a noise")
    label,freq=np.unique(data['visited'],return_counts=True)
    
```
