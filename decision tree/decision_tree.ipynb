```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
data=pd.read_csv("cardata.txt",header=None,sep=',')
X=data.values[:,0:6]
Y=data.values[:,6]
Buying=np.unique(data[0])
Maint=np.unique(data[1])
Doors=np.unique(data[2])
Persons=np.unique(data[3])
Lugboot=np.unique(data[4])
Safety=np.unique(data[5])
classlabel=np.unique(data[6])
a=list(data.loc[2,:])
print(a)
```

```python
##decision tree using my own BHAGYA(my library), ;)

def cal_entropy(arr):
    label,freq=np.unique(arr,return_counts=True)
    print(label,freq)
    total_freq=np.sum(freq)
    temp=freq/total_freq
    temp=temp*np.log2(temp)
    entropy=-1*np.sum(temp)
    return entropy,freq


clsentropy,clsfreq=cal_entropy(Y)
print("class entropy:",clsentropy)
print("class frequency:",clsfreq)
total_classfreq=np.sum(clsfreq)

def cal_attentropy(att,targetclass,total_classfreq,data):
    buyingclass=data[[att,targetclass]]
    first=data.values[:,att]
    second=data.values[:,targetclass]
    row=np.unique(first)
    column=np.unique(second)
    combination=[]
    for i in range(0,len(buyingclass)):
        a=buyingclass[att][i]+buyingclass[targetclass][i]
        combination.append(a)
    combination=np.array(combination)
    label,freq=np.unique(combination,return_counts=True)
    print(label,freq)
    total_freq=np.sum(freq)
    mat=[]
    for i in row:
        temp=[]
        for j in column:
            a=i+j
            if(a in label):
                temp.append(freq[list(label).index(a)])
            else:
                temp.append(0)
        mat.append(temp)
    #print(mat)
    att_info=[]
    att_entropy=[]
    for i in range(0,len(mat)):
        if(0 in mat[i]):
            att_info.append(0)
            continue
        total=np.sum(mat[i])
        mat[i]=mat[i]/total
        mat[i]=mat[i]*np.log2(mat[i])
        total1=np.sum(mat[i])
        att_info.append(total1*(-1))
        temp=total/total_classfreq
        temp=temp*(-1)*total1
        att_entropy.append(temp)
    att_entropy=np.sum(att_entropy)
    return att_entropy,freq


att_entropy,freq =cal_attentropy(0,6,total_classfreq,data)
#print("entropy for attribute buying:",att_entropy)
print("information gain for attribute buying=",clsentropy-att_entropy)

att_entropy,freq=cal_attentropy(1,6,total_classfreq,data)
#print("entropy for attribute Maint:",att_entropy)
print("information gain for attribute Maint=",clsentropy-att_entropy)

att_entropy,freq=cal_attentropy(2,6,total_classfreq,data)
#print("entropy for attribute Doors:",att_entropy)
print("information gain for attribute Doors=",clsentropy-att_entropy)

att_entropy,freq=cal_attentropy(3,6,total_classfreq,data)
#print("entropy for attribute Persons:",att_entropy)
print("information gain for attribute Persons=",clsentropy-att_entropy)

att_entropy,freq=cal_attentropy(4,6,total_classfreq,data)
#print("entropy for attribute lugboots:",att_entropy)
print("information gain for attribute lugboots=",clsentropy-att_entropy)

att_entropy,freq=cal_attentropy(5,6,total_classfreq,data)
#print("entropy for attribute Safety:",att_entropy)
print("information gain for attribute safety=",clsentropy-att_entropy,"which is maximum, hence it will be root")
        
```

```python
def furthercal(feature,arr,index,newdata):
    print("\n#####this is for",feature,"#####\n")
    datacollection=[]
    for attribute in feature:
        print("\n")
        print("******THIS IS FOR",attribute,"*******")
        Xall=newdata.values[:,0:7]
        newtable=[]
        newclass=[]
        for i in range(0,len(newdata)):
            if(newdata[index][i]==attribute):
                newtable.append(list(newdata.loc[i,:]))
                newclass.append(list(newdata.loc[i,6:7]))
        newtable=pd.DataFrame(data=newtable)
        newclass=np.array(newclass)

        clsentropy,clsfreq=cal_entropy(newclass)
        #print("class entropy:",clsentropy)
        #print("class frequency:",clsfreq)
        total_classfreq=np.sum(clsfreq)

        if(clsentropy==-0.0):
            print(attribute,"can't be classified further")
            continue
            
        for x in arr:
            att_entropy,freq =cal_attentropy(x,6,total_classfreq,newtable)
            print("information gain for attribute",x,"=",clsentropy-att_entropy)
        
        datacollection.append(newtable)
    return datacollection

##this is under safety bhagya feature
newtable=furthercal(Safety,[0,1,2,3,4],5,data)
print(newtable)

##this is under safety->lugboot bhagya feature
newtable1=furthercal(Lugboot,[0,1,2,3],4,pd.DataFrame(newtable[0]))
print(newtable1)

##this is under safety->Maint bhagya feature 
newtable2=furthercal(Maint,[0,2,3,4],1,pd.DataFrame(newtable[1]))
print(newtable2)
```
